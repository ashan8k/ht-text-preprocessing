```python
print(f"Pub dict has {len(pub)} items")

clean_pub_dict = {}

for k,v in pub_date.items():
    try:
        k = int(float(k))
    except ValueError:
        # Handle the exception
        print(f"Item with key {k} was removed from dictionary")
        continue

    if k >= 1800 and k <= 1899:
        clean_pub_dict[k] = v
    else:
        continue

print(f"Pub dict has {len(clean_pub_dict)} items")
```

```python
with open('bk-1800-1899-pub-dates.txt', 'w') as f:
    for k, v in sorted(clean_pub_dict.items()):
        f.write(str(k)+','+str(v)+'\n')
        
with open('bk-1800-1899-lang.txt', 'w') as g:
    for k, v in sorted(lang.items()):
        g.write(str(k)+','+str(v)+'\n')
        
with open('bk-1800-1899-rights.txt', 'w') as h:
    for k, v in sorted(rights.items()):
        if isinstance(k, str) == True:
            h.write(str(k)+','+str(v)+'\n')
        else:
            print(f"Rights value of {k} is not a string.")
```
